# Create adjacency list and graphviz dump.

```python
# Read file into lines: each line corresponds to one entry in the adjacency list which will be used later.
def parse_txt(filename):
    with open(filename) as file:
        lines = [line.rstrip() for line in file]
        return lines
    
# Convert array of lines into adjacency list.
# Using dictionary is most convenient here with node_id as key.
def build_adjlist(lines):
    adjlist = dict()
    for line in lines:
        arr = line.split(" ")
        node = arr[0]
        del arr[0]
        adjlist[node] = set(arr)
        
    return adjlist

# Print(dict) does not separate nodes into lines.
def dump_adjlist(adjlist):
    for key, value in adjlist.items():
        print(f"{key}: {value}")

def graph_dump(adjlist):
    dot = gv.Digraph()
    for node, successors in adjlist.items():
        dot.node(node, label = str(node))
        for succ in successors:
            # Graphviz does not create a new instance of the same node.
            dot.node(succ, label = str(succ))
            dot.edge(node, succ)

    return dot
```

```python
# Uncomment for sanity check for cyclic graphs.
# lines = parse_txt("cycle-graph.txt")
lines = parse_txt("graphs.txt")

adjlist = build_adjlist(lines)
graph_dump(adjlist)
```

# Add 'start' and 'end' nodes.

'End' node is attached to the leaves of the original graph, while 'start' node is attached to the leaves of the inversed graph. 

```python
def find_leaves(adjlist):
    leaves = []
    for nodes, successors in adjlist.items():
        for succ in successors:
            if succ not in adjlist:
                if succ not in leaves:
                    leaves.append(succ)

    return leaves

def build_inv_graph(adjlist):
    inverse_adjlist = dict()

    # Not optimized, but key must be declared first to add anything to it. 
    for node, successors in adjlist.items():
        for succ in successors:
            inverse_adjlist[succ] = set()

    for node, successors in adjlist.items():
        for succ in successors:
            inverse_adjlist[succ].add(node)

    return inverse_adjlist
```

```python
# Sanity check.
find_leaves(adjlist)
```

```python
def add_start_end(adjlist):
    se_adjlist = adjlist.copy()

    for leaf in find_leaves(se_adjlist):
        se_adjlist[leaf] = set()
        se_adjlist[leaf].add('end')

    se_adjlist['start'] = set()
    for inv_leaf in find_leaves(build_inv_graph(se_adjlist)):
        se_adjlist['start'].add(inv_leaf)

    return se_adjlist
```

```python
se_adjlist = add_start_end(adjlist) 
graph_dump(se_adjlist)
```

# Topological sorting and cyclic graphs.

It is convenient to check if graph is cyclic using the topological sort, not the other way around. However, the algorighm heavily relies on having the 'start' node, so it cannot resolve certain type of cycles.

In this implementation the depth-first search algorithm was used.

```python
white = 0
gray = 1
black = 2
```

```python
def getList(dict):
    arr = []
    for key in dict.keys():
        arr.append(key)
        
    return arr

def topological_sort(adjlist):
    # Stack to store sorted keys.
    sorted_keys = []
    
     # Dictionary to store information about node status
    status = dict()
    keys = getList(adjlist)
    for key in keys:
        status[key] = white
    status['end'] = white
        
    def visit(node):
        if (status[node] == black):
            return
        if (node == 'end'):
            return
            status['end'] = black
            sorted_keys.insert(0, node)
            return
        if (status[node] == gray):
            raise Exception("Graph is not acyclic")

        status[node] = gray

        for succ in adjlist[node]:
            visit(succ)

        status[node] = black
        sorted_keys.insert(0, node)

    visit('start')

    # Sorted graph.
    sort = dict()
    
    # Copy nodes.
    for node in range(len(sorted_keys)):
        sort[str(node)] = adjlist[sorted_keys[node]]

    # Substitute successors.
    for node, succs in sort.items():
            new_succs = set()
            for s in succs:
                if (s != 'end'):
                    new_succs.add(str(sorted_keys.index(s)))

            sort[node] = new_succs
    del sort['0']
    
    return sorted_keys, sort
```

# Dominators and postdominators

### For detecting dominators the following algorithm was used:
- if in the *reversed graph* `node0` has only one successor(`node1`), then in the original graph `node0` is dominated by `node1`
- if all predeccessors of `node0` have common dominators `doms`, then `doms` dominate over `node0`

### For postdominators the following quility was used:
- postdominators of `node0` are dominators of the same node in the inversed graph 

```python
def intersection(lst1, lst2):
    # Because 'start' is a dominant for every other node in the 
    # graph, we guarantee that lst2 will not be epmty after the
    # first intersection with dominators.
    if not lst2:
        return lst1
        
    lst3 = [value for value in lst1 if value in lst2]
    return lst3

# Return dominators EXCEPT for the node itself. 
def dominators(se_adjlist, node):
    inv = build_inv_graph(se_adjlist)
    doms = []
    if (len(inv[node]) == 1):
        if inv[node] not in doms:
            doms.append(inv[node])

    succ_doms = []
    for s in inv[node]:
        if (s != 'start'):
            succ_doms = intersection(dominators(se_adjlist, s), succ_doms)
    
    for d in succ_doms:
        if d not in doms:
            doms.append(d)

    return doms

def build_dom_tree(adjlist):
    se_adjlist = add_start_end(adjlist)
    dom_tree = dict()
    for node, succs in se_adjlist.items():
        if (node != 'start'):
            dom_tree[node] = set()
    
    for node, succs in se_adjlist.items():
        if (node != 'start'):
            [dom] = dominators(se_adjlist, node)[0]
            if (dom != 'start'):
                dom_tree[dom].add(node)

    return dom_tree
```

```python
dom_tree = build_dom_tree(adjlist)
graph_dump(dom_tree)
```

```python
se_inv_adjlist = build_inv_graph(adjlist)
postdom_tree = build_dom_tree(se_inv_adjlist)
graph_dump(postdom_tree)
```
